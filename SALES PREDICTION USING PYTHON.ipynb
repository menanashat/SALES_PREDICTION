```python
traing_data=sales[["TV","Radio",'Newspaper']]
testing_data=sales["Sales"]
x_train,x_test,y_train,y_test=train_test_split(traing_data,testing_data,test_size=0.2,random_state=42)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
prediction=reg.predict(x_test)
```
